```python
import pandas as pd
def duplicate_Meta_data(path):
    df = pd.read_csv(path, sep='|')
    mask = (df['meta_data'].astype(str).str.split('|')
        .explode()
        .duplicated(keep=False)
        .groupby(level=0).max())
    dd=df[mask]
    if len(dd)==0:
        print("No duplication found On Meta-Data")
    else:    
        print("...................................")
        print("Duplication from Meta-Data column")
        print(dd)


def duplicate_Variant(path):
    df = pd.read_csv(path, sep='|')
    mask = (df['variant'].astype(str).str.split('|')
        .explode()
        .duplicated(keep=False)
        .groupby(level=0).max())
    dd=df[mask]
    if len(dd)==0:
        print("No duplication found on Variant")
    else:
        print("...................................")
        print("Duplication from variant column")
        print(dd)

def space_check_Meta_data(path):
    df = pd.read_csv(path, sep='|')
    mask = (df['meta_data'])
    for i in mask:
        if " " in i:
            print("..................................")
            print("Warning: meta_data contains a space")
            print(i)

def pip_check(path):
    df = pd.read_csv(path, sep='|')
    mask=df['variant'].str.endswith('|').fillna(False)
    print("..........................")
    print("Warning: last variant is missing")
    print(df[mask])        

def Blank_or_empty_value(path):
    df = pd.read_csv(path, sep='|')
    nan_values = df[df.isna().any(axis=1)] # to get all rows with Na
    print(".........................................")
    print("Warning: variant is empty")
    print(nan_values)   
            
if __name__ == "__main__":
    path = 'D:\sample.psv'
    duplicate_Meta_data(path)
    duplicate_Variant(path)
    space_check_Meta_data(path)
    pip_check(path)
    Blank_or_empty_value(path)
```
